```python
import torch
```

```python
class myowngrad(torch.autograd.Function):
  @staticmethod
  def forward(ctx,x):
    res = x**3
    ctx.save_for_backward(x)
    return res
  @staticmethod
  def backward(ctx,grad_output):
    x, = ctx.saved_tensors
    grad_output = 3*(x**2)
    return grad_output
```

```python
x = torch.tensor(2.0,requires_grad=True)
g = myowngrad.apply
y=g(x)
print(y.grad_fn)
y.backward()
x.grad
```
