```python
import torch.nn as nn
import torch
import torch.optim as optim
```

```python
class LR(nn.Module):
  def __init__(self,input_size,output_size):
    super(LR,self).__init__()
    self.linear = nn.Linear(input_size,output_size)
  def forward(self,x):
    out = self.linear(x)
    return out
```

```python
from torch.utils.data import Dataset, DataLoader

class Data2D(Dataset):
    def __init__(self):
        self.x = torch.zeros(20, 2)
        self.x[:, 0] = torch.arange(-1, 1, 0.1)
        self.x[:, 1] = torch.arange(-1, 1, 0.1)
        self.w = torch.tensor([[2.0], [1.0]])
        self.b = 1
        self.f = torch.mm(self.x, self.w) + self.b
        self.y = self.f + 0.01 * torch.randn((self.x.shape[0], 1))
        self.len = self.x.shape[0]
    def __getitem__(self, index):
        return self.x[index], self.y[index]

    def __len__(self):
        return self.len

dataset = Data2D()
criterion = nn.MSELoss()
trainloader = DataLoader(dataset=dataset, batch_size=2)
model = LR(input_size=2, output_size=1)
optimizer = optim.SGD(model.parameters(), lr=0.1)

for epoch in range(100):
    for x, y in trainloader:
        yhat = model(x)
        loss = criterion(yhat, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

linear Regression Multiple outputs

```python
from torch.utils.data import Dataset
import torch

class Data2DD(Dataset):
    def __init__(self):
        self.x = torch.zeros(20, 2)
        self.x[:, 0] = torch.arange(-1, 1, 0.1)
        self.x[:, 1] = torch.arange(-1, 1, 0.1)

        # Two output weights for generating 2D targets
        self.w = torch.tensor([[2.0, 1.0],   # shape: (2, 2)
                               [1.0, 2.0]])
        self.b = torch.tensor([1.0, 1.0])     # shape: (2,)

        # Linear transformation with 2D output
        self.f = torch.mm(self.x, self.w) + self.b  # shape: (20, 2)

        # Add small noise to the targets
        self.y = self.f + 0.01 * torch.randn((self.x.shape[0], 2))

        self.len = self.x.shape[0]

    def __getitem__(self, index):
        return self.x[index], self.y[index]

    def __len__(self):
        return self.len
```

```python


dataset_1 = Data2DD()
criterion = nn.MSELoss()
trainloader = DataLoader(dataset=dataset_1, batch_size=2)
model_1 = LR(input_size=2, output_size=2)
optimizer = optim.SGD(model_1.parameters(), lr=0.1)

for epoch in range(100):
    for x, y in trainloader:
        yhat = model_1(x)
        loss = criterion(yhat, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
